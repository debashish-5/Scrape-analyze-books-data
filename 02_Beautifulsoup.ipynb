# DATA ANALYSIS(Storing all pages,using PANDAS,using MATPLOTLIB AND SEABORN)


```python
# Storing all page Data into items
from bs4 import BeautifulSoup

items = [] 

for i in range(1,51):
    with open(f"htmls/page{i}.html","r",encoding="utf-8")as f:
        content = f.read()

    soup = BeautifulSoup(content,"html.parser")

    articles = soup.select("article.product_pod")
    print(f"page{i} has {len(articles)} articles")

    for article in articles:
        title = article.find("h3").find("a")["title"]
        price = article.select_one("p.price_color").text.split("£")[1]
        rating_element = article.select_one("p.star-rating")
        rating = rating_element['class'][1]
        items.append([title,price,rating])
    
```

```python
print(f"\n Total books scraped:{len(items)}")
```

# Data Visualization using Matplotlib and Seaborn

## Plotting charts using Matplotlib

### What's the trend between price and reviews ?

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("ggplot")
plt.plot("Price","Rating",color = "green",data=df)
plt.xlabel("Price")
plt.ylabel("Rating")
plt.title("PRODUCT VALUE INSIGHT")
plt.grid(True)
plt.tight_layout()
```

## Bar charts in Matplotlib

### What are the top 10 highest rated books ?

```python
top_rated = df.sort_values("Rating",ascending=False).head(10)
plt.style.use("ggplot")
plt.bar(top_rated["Book Title"],top_rated["Price"],color = "red")
plt.xticks(rotation=90)
plt.xlabel("Book Title")
plt.ylabel("Price")
plt.title("TOP 10 BOOKS")
plt.grid(True)
```

## Pie charts in Mat

### What share of price do top books hold ?

```python
top5 = df.sort_values("Price",ascending=False).head(5)
top5.set_index("Book Title")['Price'].plot.pie(autopct="%1.1f%%",startangle=140,shadow=True,colors=plt.cm.Paired.colors)

plt.title("Which books hold most price Share ?")
```

### How does price ranking relate to actual prices ?

```python
plt.stackplot("Price","Price Rank",labels="Rating",data=df,colors=plt.cm.Paired.colors,edgecolor="black")
plt.legend()
plt.xlabel("Price")
plt.ylabel("Price Rank")
plt.title("Price vs Price Rank Over Books")
plt.grid(True,linestyle="--")
```

## Scatter Plots in Matplotlib

### How Does Book Price Relate to User Ratings Across Different Price Categories?

```python

rating_map = {'★': 1, '★★': 2, '★★★': 3, '★★★★': 4, '★★★★★': 5}
df['Rating_Num'] = df['Rating'].map(rating_map)


rank_map = {'Low': 1, 'Medium': 2, 'High': 3}
df['PriceRank_Num'] = df['Price Rank'].map(rank_map)


df['Group'] = pd.cut(df['PriceRank_Num'], bins=[0, 1, 2, 4], labels=['LOW', 'MEDIUM', 'HIGH'])

for group in df['Group'].unique():
    subset = df[df['Group'] == group]
    plt.scatter(subset['Price'], subset['Rating_Num'], label=group, alpha=0.6)

# Labels 
plt.xlabel('Price')
plt.ylabel('Rating')
plt.title('Price vs Rating by Group')
plt.legend(title='Group')
plt.grid(True)
plt.show()


                     
```

## Subplots in Matpotlib

### What patterns exist between Price ,Rating and Price Rank ? 

```python
plt.figure(figsize=(8,6))

plt.subplot(1,2,1)
plt.plot("Price","Rating",color="purple",data=df)
plt.xlabel("Price")
plt.ylabel("Rating")
plt.title("Price Vs Rating")

plt.subplot(1,2,2)
plt.plot("Price","Price Rank",color="limegreen",data=df)
plt.xlabel("Price")
plt.ylabel("Price Rank")
plt.tight_layout()
plt.title("Price Vs Price Rank")
```

## Seaborn in Matplotlib

```python
import seaborn as sns
df["Rating Num"]=df["Rating"].apply(lambda x:len(str(x).strip()) if isinstance(x,str) else 0)

df = df.dropna(subset=["Rating","Price Rank"])
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='Rating',hue='Price Rank',palette="Set2")
plt.title("Book Rating Distribution by Price Rank")
plt.xlabel("Rating")
plt.ylabel("Number of Books")
plt.legend(title="Price Rank")
plt.tight_layout()

```
